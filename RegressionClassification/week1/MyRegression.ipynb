```python
import numpy as np
import matplotlib.pyplot as plt
```

To generate the training dataset, let's take the $y = kx + b$ function and add some randomness.

```python
rng = np.random.default_rng()

# Generate sample data: x from 1 to 100 with step of 1
x_train = np.arange(1, 100, 1)
m = x_train.size
# Add some randomness to the value of Y
y_train = (7 * x_train + 650) + (50 - 100 * rng.random(m))

# Get predicted result for given w, b and x
def predict(w, b, x):
    return w * x + b

# Get partial derivatives dJ/dw and dJ/db
def derivative(w, b, x, y):
    dj_dw = 0;
    dj_db = 0;
    for i in range(m):
        predicted = predict(w, b, x[i])
        diff = predicted - y[i]
        dj_dw = dj_dw + diff * x[i]
        dj_db = dj_db + diff

    return dj_dw/m, dj_db/m

# Set gradient descent parameters
alpha = 1 * (10 ** (-4))
steps_count = 1 * (10 ** 6)

# Set initial w and b
w_current = 0
b_current = 500

# Run gradient descent
for i in range(steps_count):
    w_tmp = w_current
    b_tmp = b_current
    dj_dw, dj_db = derivative(w_tmp, b_tmp, x_train, y_train)
    w_current= w_tmp - alpha * dj_dw
    b_current = b_tmp - alpha * dj_db


# Print result and compare given and predicted f(x) plots
print(f'f = {w_current} * x + {b_current}')

plt.scatter(x_train, y_train, marker='x', c='r')
plt.plot(x_train, predict(w_current, b_current, x_train))
plt.show()
```
